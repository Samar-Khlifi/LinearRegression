<h2><font color='#mmvb'>Exercice n° 1:</font></h2>

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
samar_khlifi = pd.read_csv('grav.csv')
print("shape:", samar_khlifi.shape)
samar_khlifi.head()
```

```python
X= samar_khlifi['height']
y= samar_khlifi['seconds squared']
plt.scatter(X ,y)
plt.xlabel("height")
plt.ylabel("seconds squared")
```

```python
np.corrcoef(X,y)
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(X,y)

print('Slope: ',slope,'\nIntercept: ',intercept)
```

```python
from scipy import stats
#linregress(): calcule la regression linéaire
#On s'interesse au slope et intercept
slope, intercept, r_value, p_value, std_err = stats.linregress(X, y)
def predict(x):
    return slope * x + intercept
#fitLine:prendre des valeurs prédites depuis X
fitLine = predict(X)
plt.plot(X, fitLine, c='y')
```

```python
from matplotlib import pyplot as plt
plt.scatter(X, fitLine)
plt.plot(X, fitLine,'y')
plt.show()
```

```python
# Import the numpy and pandas package, pyreadstat
import numpy as np
import pandas as pd
import pyreadstat
# Read the given sav file, and view some sample records
samar_khlifi, meta = pyreadstat.read_sav('test_scores.sav')
samar_khlifi
```

```python
# Shape of our dataset
samar_khlifi.shape

# Info our dataset
samar_khlifi.info()

# Describe our dataset
samar_khlifi.describe()
```

```python
# Import matplotlib and seaborn libraries to visualize the data
import matplotlib.pyplot as plt 
import seaborn as sns

# Using pairplot we'll visualize the data for correlation
sns.pairplot(samar_khlifi, x_vars=['school','school_setting','school_type','classroom','teaching_method','n_student','student_id','gender', 'lunch','pretest'], 
             y_vars='posttest', size=4, aspect=1, kind='scatter')
plt.show()
```

```python
# Creating X and y
X = samar_khlifi['pretest']
y = samar_khlifi['posttest']
plt.scatter(X,y)
plt.xlabel("pretest")
plt.ylabel("posttest")
```

```python
!pip install statsmodels==0.10.0rc2 --pre --user
```

```python
# Splitting the varaibles as training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                    test_size = 0.3, random_state = 100)
```

```python
# Take a look at the train dataset
X_train
y_train
```

```python
# Importing Statsmodels.api library from Stamodel package
import statsmodels.api as sm
# Adding a constant to get an intercept
X_train_sm = sm.add_constant(X_train)
# Fitting the resgression line using 'OLS'
lr = sm.OLS(y_train, X_train_sm).fit()
# Printing the parameters
lr.params
```

```python
# Performing a summary to list out all the different parameters of the regression line fitted
lr.summary()
```

```python
# Visualizing the regression line
plt.scatter(X_train, y_train)
plt.plot(X_train, 13.3750 + 0.9783*X_train, 'y')
plt.show()
```

```python
# Predicting y_value using traingn data of X
y_train_pred = lr.predict(X_train_sm)
# Creating residuals from the y_train data and predicted y_data
res = (y_train - y_train_pred)
# Plotting the histogram using the residual values
fig = plt.figure()
sns.distplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()
```

```python
# Looking for any patterns in the residuals
plt.scatter(X_train,res)
plt.show()
```

```python
# Adding a constant to X_test
X_test_sm = sm.add_constant(X_test)
# Predicting the y values corresponding to X_test_sm
y_test_pred = lr.predict(X_test_sm)
# Printing the first 15 predicted values
y_test_pred
```

```python
# Importing r2_square
from sklearn.metrics import r2_score
# Checking the R-squared value
r_squared = r2_score(y_test, y_test_pred)
r_squared
```

```python
# Visualize the line on the test set
plt.scatter(X_test, y_test)
plt.plot(X_test, y_test_pred, 'y')
plt.show()
```

```python
# Splitting the data into train and test
from sklearn.model_selection import train_test_split
X_train_lm, X_test_lm, y_train_lm, y_test_lm = train_test_split(X, y, train_size = 0.7, 
                                                                test_size = 0.3, random_state = 100)

# Shape of the train set without adding column
X_train_lm.shape
# Adding additional column to the train and test data
X_train_lm = X_train_lm.values.reshape(-1,1)
X_test_lm = X_test_lm.values.reshape(-1,1)
print(X_train_lm.shape)
print(X_test_lm.shape)
```

```python
from sklearn.linear_model import LinearRegression
# Creating an object of Linear Regression
lm = LinearRegression()
# Fit the model using .fit() method
lm.fit(X_train_lm, y_train_lm)
# Intercept value
print("Intercept :",lm.intercept_)
# Slope value
print('Slope :',lm.coef_)
```

```python
# Making Predictions of y_value
y_train_pred = lm.predict(X_train_lm)
y_test_pred = lm.predict(X_test_lm)
# Comparing the r2 value of both train and test data
print(r2_score(y_train,y_train_pred))
print(r2_score(y_test,y_test_pred))
```

```python
#plot the distribution of the target variable MEDV
sns.set()
sns.distplot(boston_dataset['MEDV'], bins=15)
plt.show()
```

```python
# Visualizing the data using heatmap
sns.heatmap(boston_dataset.corr().round(2), cmap="YlGnBu", annot = True)
plt.show()
```

```python
plt.figure()
features = ['LSTAT', 'RM']
target = boston_dataset['MEDV']
for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = boston_dataset[col]
    y = target
    plt.scatter(x, y)
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
```

```python
X = pd.DataFrame(np.c_[boston_dataset['LSTAT'], boston_dataset['RM']], columns = ['LSTAT','RM'])
y = boston_dataset['MEDV']
```

```python
# Splitting the data into train and test
from sklearn.model_selection import train_test_split
X_train_lm, X_test_lm, y_train_lm, y_test_lm = train_test_split(X, y, train_size = 0.7, 
                                                                test_size = 0.3, random_state = 100)

# Shape of the train set without adding column
X_train_lm.shape
# Adding additional column to the train and test data
X_train_lm = X_train_lm.values.reshape(-1,1)
X_test_lm = X_test_lm.values.reshape(-1,1)
print(X_train_lm.shape)
print(X_test_lm.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
# model evaluation for training set
y_train_predict = lm.predict(X_train)
rmse = (np.sqrt(mean_squared_error(y_train, y_train_predict)))
r2 = r2_score(y_train, y_train_predict)

print("The model performance for training set:")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
# model evaluation for testing set
y_test_predict = lm.predict(X_test)
rmse = (np.sqrt(mean_squared_error(y_test, y_test_predict)))
r2 = r2_score(y_test, y_test_predict)
print("The model performance for testing set:")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
